## The CORES dataset text file claims that each example's IQ is normalized to have unity average magnitude but this is not true

```python
magnitude = [np.sqrt(x[0][i]**2 + x[1][i]**2) for i in range(256)]
mean_mag = np.mean(magnitude)
mean_mag
```

```python
normalize_factor = 1/mean_mag
normalize_factor
```

```python
normalized_x = x*normalize_factor
normalized_x
```

```python
x = normalized_x
magnitude = [np.sqrt(x[0][i]**2 + x[1][i]**2) for i in range(256)]
mean_mag = np.mean(magnitude)
mean_mag
```
